# Data Preparation

```python
new_data.columns
```

```python
new_data[['adults','children','babies']]
```

There are some rows with total number of adults, children or babies equal to zero. So we will merge it.

```python
#Lets merge adults, children and babies together as kids
new_data['total_members'] = new_data.children + new_data.babies + new_data.adults
```

```python
new_data.info()
```

```python
#convert the datatypes to string
new_data[['children', 'adr']] = new_data[['children', 'adr']].astype('int64')
new_data['total_members'] = new_data['total_members'].astype('int64')
```

```python
#creating some new features as total_nights
new_data["total_nights"] = new_data['stays_in_weekend_nights'] + new_data['stays_in_week_nights']


# creating the new feature as guest_type 
lists = []
for i in new_data['total_members']:
    if i == 1:
        lists.append('single')
    elif i == 2:
        lists.append('couple')
    else:
        lists.append('family')
new_data['Guest_type'] = lists
```

```python
# Dropping the columns which is unwanted
new_data.drop(['babies','previous_bookings_not_canceled','days_in_waiting_list','total_of_special_requests'
               ,'assigned_room_type','reservation_status_date','booking_changes'],axis = 1,inplace=True)
```

```python
new_data.hotel.value_counts()
```

```python
# writing a function to Notate the percent count of each value on the bars
def annot_percent(axes):
    for p in plot.patches:
        total = sum(p.get_height() for p in plot.patches)/100
        percent = round((p.get_height()/total),2)
        x = p.get_x() + p.get_width()/2
        y = p.get_height()
        plot.annotate(f'{percent}%', (x, y), ha='center', va='bottom')
```

```python
plt.figure(figsize=(8,8))
plt.title("Hotel")
plot = sns.countplot('hotel',data=new_data)
annot_percent(plot)
```

### Observation:
We can see by the above, that mojority of the people preferred city hotel over resort hotel.
Around 66% bookings are for City hotel and 33% bookings are for Resort hotel.
